```python
# Import Python libraries
import pandas as pd
import numpy as np
```

```python
# Read CSV
apps_with_duplicates = pd.read_csv('datasets/apps.csv')

print(apps_with_duplicates.head())
```

```python
# Cleaning the DataFrame
apps = apps_with_duplicates.drop_duplicates()
#print(apps.shape)
#print(apps.info())

# List of characters to remove
chars_to_remove = ['+', ',']
# List of column names to clean
cols_to_clean = 'Installs'

# Loop for each char in chars_to_remove
for char in chars_to_remove:
    # Replace the character with an empty string
    apps['Installs'] = apps['Installs'].apply(lambda x: x.replace(char,''))
        
# Print a summary of the apps dataframe
#print(apps.info())

# Convert the column 'Installs' into integer data type
apps['Installs'] = apps['Installs'].astype(np.int64)

print(apps.info())
```

```python
# Group by different categories
app_category_info = apps.groupby('Category')
#print(app_category_info.head())

# Find the number of apps in each category, the average price and the average rating
app_category_info = apps.groupby('Category').agg({'App': 'count', 'Price': 'mean', 'Rating': 'mean'})
#print(app_category_info.head())

# Rename the columns for easier understanding
app_category_info = app_category_info.rename(columns={"App": "Number of apps", "Price": "Average price", "Rating": "Average rating"})
print(app_category_info.head())
```

```python
#print(apps.head())
# Find the top10 free FINANCE apps having the highest average sentiment score
df_free_finance_apps = apps[(apps['Type'] == 'Free') & (apps['Category'] == 'FINANCE')]
print(df_free_finance_apps.head())
```

```python
# Load datasets/user_reviews.csv
user_reviews = pd.read_csv('datasets/user_reviews.csv')
print(user_reviews.head())
```

```python
# Merge df_free_finance_apps and user_reviews dataframes
top_10_user_feedback = df_free_finance_apps.merge(user_reviews, on='App')
print(top_10_user_feedback.head())
```

```python
# Group by App and calculate the average of the Sentiment Score column 
top_10_user_feedback = top_10_user_feedback.groupby('App').agg({'Sentiment Score':'mean'})
#print(top_10_user_feedback.head())
```

```python
# Sort the top_10_user_feedback and slicing to get only the 10 highest scores
top_10_user_feedback = top_10_user_feedback.sort_values(by='Sentiment Score', ascending=False)
top_10_user_feedback = top_10_user_feedback[0:10]
print(top_10_user_feedback.head())
```
