```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt

# Import the data from the CSV file
office_episodes = pd.read_csv('datasets/office_episodes.csv')

#print(office_episodes.head())
#print(office_episodes.info())
```

```python
# Generating the color scheme
color = []
for ind, row in office_episodes.iterrows():
    if row['scaled_ratings'] < 0.25:
        color.append('red')
    elif row['scaled_ratings'] < 0.5:
        color.append('orange')
    elif row['scaled_ratings'] < 0.75:
        color.append('lightgreen')
    else:
        color.append('darkgreen')
```

```python
# Setting a sizing system for ep. that have guests or not
sizing_syst = []
for ind, row in office_episodes.iterrows():
    if row['has_guests'] == True:
        sizing_syst.append(250)
    else:
        sizing_syst.append(25)
```

```python
fig = plt.figure()
plt.rcParams['figure.figsize'] = [11, 7]
plt.scatter(office_episodes['episode_number'],
                     office_episodes['viewership_mil'],
                     color=color,
                     s=sizing_syst)
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.grid(); plt.show()
```
